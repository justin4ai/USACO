````python
# Solution
n = 14
ordering = "GG HG GH HG HH HG HG"

answer = 0
def photoshoot(n, ordering):

    for j in range(n - 2, -1, -2): # 뒤에서부터, 홀수번째(인덱스 상으론 짝수지만) 자리를 돌아주는 forloop
        
        if (ordering[j] != ordering[j + 1]) and ( (ordering[j] == 'G')  == (answer % 2 == 0) )   # 1. 인접한 두 글자가 서로 달라야한다.
                                                                                            # 2. (홀수번째에 G가 있는 경우 + answer가 짝수인 경우) or (짝수번째에 G가 있는 경우 + answer가 홀수인 경우)
            answer += 1

    return answer


print(answer)
````
